```python

# ğŸ¥ Step 1: Download YouTube Audio (yt-dlp approach)
import os
import subprocess

def download_audio_ytdlp(youtube_url, filename="audio"):
    output_file = f"{filename}.mp3"
    command = [
        "yt-dlp",
        "-x", "--audio-format", "mp3",
        "-o", f"{filename}.%(ext)s",
        youtube_url
    ]
    try:
        subprocess.run(command, check=True)
        if os.path.exists(output_file):
            print(f"Downloaded and converted to MP3: {output_file}")
            return output_file
        else:
            raise Exception("Download failed or file not found.")
    except subprocess.CalledProcessError as e:
        print("yt-dlp failed:", e)
        raise

# Example YouTube URL
video_url = "https://m.youtube.com/watch?v=0O2Rq4HJBxw"
#video_url = "https://www.youtube.com/watch?v=dQw4w9WgXcQ"
file_name = "youtube_audio"
audio_path = download_audio_ytdlp(video_url, file_name)
```

```python
audio_path = '/content/youtube_audio.mp3'
```

```python
# ğŸ—£ï¸ Step 2: Transcribe Audio with Whisper
import whisper
import os

def transcribe_audio(audio_file, model_size="base"):
    model = whisper.load_model(model_size)
    if os.path.exists(audio_file):
        result = model.transcribe(audio_file)
        transcript = result["text"]
        with open(audio_file + ".txt", "w") as f:
            f.write(transcript)
        return transcript
    else:
        raise FileNotFoundError(f"{audio_file} not found")

transcript = transcribe_audio(audio_path)
print(transcript[:10000000])  # preview
```

```python
import torch
```

```python
# ğŸ”„ Step 3: Load Hugging Face summarization model
from transformers import pipeline

# Load summarizer pipeline (uses facebook/bart-large-cnn by default)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=0 if torch.cuda.is_available() else -1)
```

```python
# ğŸ§  Step 4: Summarize Using Hugging Face Model

# Split transcript if too long for model
max_chunk_size = 1024
chunks = [transcript[i:i+max_chunk_size] for i in range(0, len(transcript), max_chunk_size)]

summary = ""
for chunk in chunks:
    result = summarizer(chunk, max_length=130, min_length=30, do_sample=False)
    summary += result[0]['summary_text'] + "\n"

print(summary)
```
